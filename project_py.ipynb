```python
# Define SQL query
query = """
WITH ProductRevenue AS (
    SELECT
        product_id,
        category,
        sub_category,
        SUM(list_price * quantity * (1 - discount_percent)) AS total_revenue,
        SUM((list_price - cost_price) * quantity * (1 - discount_percent)) AS total_profit,
        COUNT(order_id) AS total_orders
    FROM store.Orders  -- Ensure schema.table reference is correct
    GROUP BY product_id, category, sub_category
),
RankedProducts AS (
    SELECT
        product_id,
        category,
        sub_category,
        total_revenue,
        total_profit,
        total_orders,
        ROW_NUMBER() OVER (ORDER BY total_revenue DESC) AS revenue_rank,
        ROW_NUMBER() OVER (ORDER BY total_profit DESC) AS profit_rank,
        CASE
            WHEN total_revenue > 100000 THEN 'High Performer'
            WHEN total_revenue BETWEEN 50000 AND 100000 THEN 'Moderate Performer'
            ELSE 'Low Performer'
        END AS performance_category
    FROM ProductRevenue
)
SELECT * FROM RankedProducts
WHERE revenue_rank <= 10 OR profit_rank <= 10
ORDER BY revenue_rank;
"""

# Execute the query
mycursor.execute(query)

# Fetch results
results = mycursor.fetchall()

# Print results
for row in results:
    print(row)

# Close connection
mycursor.close()
connection.close()
```

```python
# ✅ Define SQL Query for Discount Analysis
query = """
SELECT
    product_id,
    category,
    SUM(list_price * quantity) AS total_sales_before_discount,
    SUM(list_price * quantity * (1 - discount_percent)) AS total_sales_after_discount,
    AVG(discount_percent) AS avg_discount
FROM store.Orders
WHERE discount_percent > 0.20
GROUP BY product_id, category
ORDER BY total_sales_after_discount DESC;
"""

# ✅ Execute Query
cursor = connection.cursor()
cursor.execute(query)

# ✅ Fetch Data and Load into DataFrame
data = cursor.fetchall()
df = pd.DataFrame(data, columns=[
    "Product ID", "Category", "Sales Before Discount",
    "Sales After Discount", "Average Discount"
])
# ✅ Close Cursor & Connection
cursor.close()
connection.close()

# ✅ Print the result
print(df)
```

```python
# ✅ SQL Query: Discount Analysis Using PK & FK
query = """
SELECT
    o.order_id,  -- 🔑 Primary Key
    o.product_id,  -- 🔗 Foreign Key
    o.category,
    SUM(o.list_price * o.quantity) AS total_sales_before_discount,
    SUM(o.list_price * o.quantity * (1 - o.discount_percent)) AS total_sales_after_discount,
    AVG(o.discount_percent) AS avg_discount
FROM store.orders o
WHERE o.discount_percent > 0.20
GROUP BY o.order_id, o.product_id, o.category
ORDER BY total_sales_after_discount DESC;
"""

# ✅ Execute Query
mycursor = connection.cursor()
mycursor.execute(query)

# ✅ Fetch Data and Load into DataFrame
data = mycursor.fetchall()
df = pd.DataFrame(data, columns=[
    "Order ID", "Product ID", "Category",
    "Sales Before Discount", "Sales After Discount", "Average Discount"
])
#closecursor & connection
mycursor.close()
connection.close()

# ✅ Print the result
print(df)
```

```python
# ✅ Join Query: Top 10 Revenue Products
query = """
SELECT
    o.product_id,
    p.product_name,
    p.category,
    SUM(o.list_price * o.quantity * (1 - o.discount_percent)) AS total_revenue
FROM store.Orders o
JOIN store.Products p ON o.product_id = p.product_id
GROUP BY o.product_id, p.product_name, p.category
ORDER BY total_revenue DESC
LIMIT 10;
"""

# ✅ Execute Query
mycursor.execute(query)

# ✅ Fetch Results
data = mycursor.fetchall()

# ✅ Convert to DataFrame
df = pd.DataFrame(data, columns=["Product ID", "Product Name", "Category", "Total Revenue"])
# ✅ Display Output
print(df)

# ✅ Close Cursor & Connection
mycursor.close()
connection.close()
```

```python
# ✅ SQL Query
query = """
SELECT
    city,
    SUM((list_price - cost_price) * quantity) AS total_profit,
    SUM(list_price * quantity * (1 - discount_percent)) AS total_revenue,
    (SUM((list_price - cost_price) * quantity) / SUM(list_price * quantity * (1 - discount_percent))) * 100 AS profit_margin
FROM store.Orders
GROUP BY city
ORDER BY profit_margin DESC
LIMIT 5;
"""

# ✅ Execute Query
mycursor.execute(query)

# ✅ Fetch Results & Convert to DataFrame
data = mycursor.fetchall()
df = pd.DataFrame(data, columns=["City", "Total Profit ($)", "Total Revenue ($)", "Profit Margin (%)"])
# ✅ Close Cursor & Connection
mycursor.close()
connection.close()

# ✅ Display Results
print(df)
```
